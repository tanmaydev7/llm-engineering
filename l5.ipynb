### Ollama Cloud 

```python
import os
import json
import gradio as gr
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
from ollama import ChatResponse
from ollama import Client

# Changes for Accessing ollama cloud hosted model
model = "gpt-oss:120b-cloud"

client = Client(
    host="https://ollama.com",
    headers={'Authorization': 'Bearer ' + os.environ.get('OLLAMA_API_KEY')}
)


system_prompt = f"""
    You are a supply chain expert and a specialize in data science and analyst.
    If someone asks you to calculate something or how to find something you provide 
    step by step process with example and formulas if possible.
    Nothing else. You are not to entertain user. You are genius.
    Make sure any formula you create, keep it Markdown compliant.
    No use of LaTeX code
""".strip()

def message_ai(prompt, history):
    clean_history = [{"role": hist.get("role"), "content": hist.get("content")} for hist in history]
    response: ChatResponse = client.chat(
        model=model, 
        messages=[
            {"role": "system", "content": system_prompt},
            *clean_history,
            {"role": "user", "content": prompt},
        ], 
        stream=True,
        think="high"
    )
    content = ""
    for chunk in response:
        content += (chunk.message.content or '')
        yield content
```

```python
gr.ChatInterface(fn=message_ai, type="messages").launch(inbrowser=True)
```
